```python
import random

# Define fish types and their prices (using the fish types from your earlier definition)
fish_prices = {'catla': 150, 'rohu': 200, 'sinchu': 250}

# Initialize individual fish counts and total earnings
catla_count = 0
rohu_count = 0
sinchu_count = 0

total_earnings = 0

print("Welcome to the Simple Fish Market Game!")

# Main game loop
while True:
    print("\n--- Menu ---")
    print("1. Go Fishing")
    print("2. Sell Fish")
    print("3. View My Catch (Inventory)")
    print("4. View Total Earnings")
    print("5. Exit Game")

    try:
        choice = int(input("Enter your choice: "))
    except ValueError:
        print("Invalid input. Please enter a number from the menu.")
        continue

    if choice == 1:  # Go Fishing
        caught_fish = random.choice(list(fish_prices.keys()))
        if caught_fish == 'catla':
            catla_count += 1
        elif caught_fish == 'rohu':
            rohu_count += 1
        elif caught_fish == 'sinchu':
            sinchu_count += 1
        print(f"You cast your net and caught a {caught_fish.capitalize()}!")

    elif choice == 2:  # Sell Fish
        # Check if there are any fish to sell
        if catla_count + rohu_count + sinchu_count == 0:
            print("You have no fish to sell. Go fishing first!")
            continue

        print("\n--- Sell Your Catch ---")
        print("Your current catch:")
        if catla_count > 0: print(f"  Catla: {catla_count} (Price: {fish_prices['catla']} coins/each)")
        if rohu_count > 0: print(f"  Rohu: {rohu_count} (Price: {fish_prices['rohu']} coins/each)")
        if sinchu_count > 0: print(f"  Sinchu: {sinchu_count} (Price: {fish_prices['sinchu']} coins/each)")

        fish_to_sell = input("Enter the type of fish you want to sell (e.g., 'catla'): ").lower()

        if fish_to_sell not in fish_prices:
            print("That's not a valid fish type.")
            continue

        current_quantity = 0
        if fish_to_sell == 'catla': current_quantity = catla_count
        elif fish_to_sell == 'rohu': current_quantity = rohu_count
        elif fish_to_sell == 'sinchu': current_quantity = sinchu_count

        if current_quantity == 0:
            print(f"You don't have any {fish_to_sell.capitalize()} to sell.")
            continue

        try:
            quantity_to_sell = int(input(f"How many {fish_to_sell.capitalize()} do you want to sell (you have {current_quantity})? "))
            if quantity_to_sell <= 0:
                print("You must sell at least one fish.")
                continue
            if quantity_to_sell > current_quantity:
                print(f"You only have {current_quantity} {fish_to_sell}(s). You can't sell that many.")
                continue

            price_per_fish = fish_prices[fish_to_sell]
            earnings_from_sale = quantity_to_sell * price_per_fish
            total_earnings += earnings_from_sale

            # Update individual fish count after selling
            if fish_to_sell == 'catla': catla_count -= quantity_to_sell
            elif fish_to_sell == 'rohu': rohu_count -= quantity_to_sell
            elif fish_to_sell == 'sinchu': sinchu_count -= quantity_to_sell

            print(f"You sold {quantity_to_sell} {fish_to_sell}(s) for {earnings_from_sale} coins!")
            print(f"Your total earnings are now {total_earnings} coins.")

        except ValueError:
            print("Invalid quantity. Please enter a whole number.")

    elif choice == 3:  # View Inventory
        print("\n--- Your Current Catch ---")
        if catla_count + rohu_count + sinchu_count == 0:
            print("Your net is empty. Time to go fishing!")
        else:
            if catla_count > 0: print(f"  Catla: {catla_count}")
            if rohu_count > 0: print(f"  Rohu: {rohu_count}")
            if sinchu_count > 0: print(f"  Sinchu: {sinchu_count}")

    elif choice == 4:  # View Total Earnings
        print(f"\n--- Total Earnings ---")
        print(f"You have a total of {total_earnings} coins.")

    elif choice == 5:  # Exit Game
        print("Thanks for playing the Fish Market Game!")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 5.")
```

```python
import random

fish_dict={'catla':12,'rohu':18,'sinchu':25}
inventory={}
total_money=0
fish_prices={'catla':12,'rohu':18,'sinchu':25}

def show_menu():
  print("\n  fishing game menu")
  print("1.catch fish")
  print("2.view inventory")
  print("3.sell fish")
  print("4.exit")

def catch_fish():
  global inventory # Declare global if modifying it directly
  fish=random.choice(list(fish_prices.keys()))
  inventory[fish]=inventory.get(fish,0)+1
  print(f"You caught a {fish}!")

def view_inventory():
  if not inventory:
    print("Your net is empty. Go fishing!")
  else:
    print("\n--- Your Inventory ---")
    for fish,count in inventory.items():
      print(f"  {fish.capitalize()}: {count}")

def sell_fish():
  global inventory, total_money
  if not inventory:
    print("No fish to sell!")
  else:
    earnings = 0
    for fish, count in inventory.items():
      earnings += fish_prices[fish] * count
    total_money += earnings
    inventory.clear()
    print(f"You sold all fish for ${earnings} coins!")
    print(f"Your total money is now ${total_money} coins.")

while True:
  try:
    choice=int(input("Enter your choice: "))
  except ValueError:
    print("Invalid input. Please enter a number.")
    continue

  if choice==1:
    catch_fish()
  elif choice==2:
    view_inventory()
  elif choice==3:
    sell_fish()
  elif choice==4:
    print(f"\nGame Over! Final money earned: ${total_money} coins.")
    print("Thanks for playing!")
    break
  else:
    print("Invalid choice. Please try again.")
  show_menu()
```
