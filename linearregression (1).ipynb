```python
import pandas as pd
import numpy as np
data=pd.DataFrame({
    'plot':['A','B','C','D','E'],
    'yield_kg':[2500,np.nan,2800,np.nan,2600],
    'rainfall':[150,200,np.nan,180,170],
    'temperature':[28,30,29,np.nan,27]
})
print("Data with missing values")
print(data)
print("Detecting missing values")
print("Is null?")
print(data.isnull())
print(f"\nTotal missing per column:")
print(data.isnull().sum())
```

```python
print("Filling strategies")

data_mean=data.copy()
print(data_mean['yield_kg'])
data_mean['yield_kg']=data_mean['yield_kg'].fillna(data_mean['yield_kg'].mean())
print('\n',"Fill with mean:")
print(data_mean['yield_kg'])
```

```python
data_median=data.copy()
print(data_median['rainfall'])
data_median['rainfall']=data_median['rainfall'].fillna(data_median['rainfall'].median())
print(data_median['rainfall'])
```

```python
data_ffill=data.copy()
print(data_ffill)
data_ffill=data_ffill.fillna(method='ffill')
print("\nForward Fill")
print(data_ffill)
```

```python
print("Dropping rows")
data_dropped=data.dropna()
print(data_dropped)
print('Original rows:',len(data))
print("After drop:",len(data_dropped))
```

```python
import pandas as pd
dataset=pd.read_csv('Salary_Data.csv')
print(dataset)
```

slice the dataset into dependent and independent variables.

```python
X=dataset.iloc[:,0:1].values
y=dataset.iloc[:,1].values
print(X)
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=1/3,random_state=0)
print(dataset.shape)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python

import numpy as np
x_mean=np.mean(X_train)
y_mean=np.mean(y_train)
print(x_mean)
print(y_mean)
```

```python
n=len(X_train)
num,den=0,0
for i in range(n):
  num+=(X_train[i]-x_mean)*(y_train[i]-y_mean)
  den+=(X_train[i]-x_mean)**2

b1=num/den
b0=y_mean-(b1*x_mean)
print("b0-",b0)
print('b1-',b1)
```

```python
y_pred_code=b0+b1*X_test
print(y_pred_code)
```

```python
print(y_test)
```

```python
n1=len(y_test)
mse=0
for i in range(n1):
  mse=(y_test[i]-y_pred_code[i])**2
msme=mse/n1
print(msme)
rmse=np.sqrt(mse)
print(rmse)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train,y_train)
```

```python
y_pred=reg.predict(X_test)
print(y_pred)
```

```python
print("Intercept",reg.intercept_)
print("Coefficient",reg.coef_)
```

evaluate metrics

```python
from sklearn import metrics
print("Mean absolute error: ")
print(metrics.mean_absolute_error(y_test,y_pred))
```

```python
from sklearn import metrics
print("Mean squared error: ",metrics.mean_squared_error(y_test,y_pred))
```

```python
print("Root mean squared error: ",np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```
