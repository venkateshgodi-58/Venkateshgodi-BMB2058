**Concatenating of two arrays.**
**If axis=0 then it concatenates along rows.**

```python
import numpy as np
arr1=np.array([[1,2],[3,4]])
arr2=np.array([[5,6],[7,8]])
concatenated_arr=np.concatenate((arr1,arr2),axis=0)
print(concatenated_arr)
```

**If axis=1 then it concatenates along columns.**

```python
import numpy as np
arr1=np.array([[1,2],[3,4]])
arr2=np.array([[5,6],[7,8]])
concatenated_arr=np.concatenate((arr1,arr2),axis=1)
print(concatenated_arr)
```

**Transpose-Interchange the rows and columns.
Shape of the matrix will also interchanged by number of rows and number of columns.**

```python
import numpy as np
arr1=np.array([[1,2,3,4],[6,7,8,9]])
transpose_arr=arr1.transpose()
print(arr1)
print(transpose_arr)
```

```python
import numpy as np
crop_yields=np.array([[12.5,13.2,14.1,13.8,15.2],
              [10.3,11.5,10.8,12.1,11.9],
              [15.1,14.8,16.2,15.5,16.8],
              [11.2,10.9,12.3,11.7,12.8],
              ])
print("Crop yield array(tons):")
print(crop_yields)
print("\nShape",crop_yields.shape)
print(f'Data type:{crop_yields.dtype}')
print("Total elements:",crop_yields.size)
print("Dimension:",crop_yields.ndim)
```

```python
total_yield=np.sum(crop_yields)
print("Total yield:",total_yield)
```

```python
median=np.median(crop_yields)
print("Median of crop yield:",median)
```

```python
sd=np.std(crop_yields)
print("Standard deviation:",sd)
```

**Axis=1--along the rows,
it gives the average among rows.**

```python
avg_yield_per_farm=np.mean(crop_yields,axis=1)
print("Mean yield per farm:",avg_yield_per_farm)
```

```python
avg_yield_per_farm=np.mean(crop_yields)
print("Mean yield per farm:",avg_yield_per_farm)
```

**Axis=0--along the columns,
it gives the average among columns.**

```python
avg_yield_per_season=np.mean(crop_yields,axis=0)
print("Mean yield per season:",avg_yield_per_season)
```

```python
best_farm=np.argmax(avg_yield_per_farm)
print("Best farm:",best_farm+1)
print('Their average yield:',round(avg_yield_per_farm[best_farm],2),"tons")
```

```python
worst_farm=np.argmin(avg_yield_per_farm)
print("Worst farm:",worst_farm+1)
print("Their average yield:",round(avg_yield_per_farm[worst_farm],1),"tons/ha")
```
