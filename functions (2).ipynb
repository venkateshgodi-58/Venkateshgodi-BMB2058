```python
inventory={}
total_money=0

fish_price={"Rohu":10,"Catla":20,"Sinchu":30}

print("🙏Welcome to Lake fishing game🙏")
print("1.Catch fishing")
print("2. View Inventory")
print("3.Sell fish💲")
print("4.exit")
while True:
  ch=int(input("Enter a choice"))

  if ch==1:
    fish=random.choice(list(fish_price.keys()))
    inventory[fish]=inventory.get(fish,0)+1
    print(f"You caught a {fish}")

  elif ch==2:
    if inventory=={}:
      print("❌ fish caught")
    else:
      print("Your inventory.")
      for fish,count in inventory.items():
        print(f"{fish}:{count}")

  elif ch==3:
    if inventory=={}:
      print("❌ fish to sell")
    else:
      earnings=0
      for fish,count in inventory.items():
         earnings+=fish_price[fish]*count
      total_money+=earnings
      inventory.clear()
      print(f"You sold all the fish for 💲{earnings}")

  elif ch==4:
      print(f"Total money earned is 💲💰{total_money}!")
      print("👏Thanks for playing")
      break

  else:
    print("Invalid input please try again")
```

```python
inventory={}
total_money=0

fish_price={"Rohu":10,"Catla":20,"Sinchu":30}

def showmenu():
  print("🙏Welcome to Lake fishing game🙏")
  print("1.Catch fishing")
  print("2. View Inventory")
  print("3.Sell fish💲")
  print("4.exit")

def catchfish():
  fish=random.choice(list(fish_price.keys()))
  inventory[fish]=inventory.get(fish,0)+1
  print(f"You caught a {fish}")

def showinventory():
  if inventory=={}:
      print("❌ fish caught")
  else:
      print("Your inventory.")
      for fish,count in inventory.items():
        print(f"{fish}:{count}")

def sellfish():
  global total_money
  if inventory=={}:
      print("❌ fish to sell")
  else:
      earnings=0
      for fish,count in inventory.items():
         earnings+=fish_price[fish]*count
      total_money+=earnings
      inventory.clear()
      print(f"You sold all the fish for 💲{earnings}")

showmenu()

while True:
  ch=int(input("Enter a choice"))

  if ch==1:
    catchfish()

  elif ch==2:
    showinventory()

  elif ch==3:
    sellfish()

  elif ch==4:
      print(f"Total money earned is 💲💰{total_money}!")
      print("👏Thanks for playing")
      break

  else:
    print("Invalid input please try again")
```

**Local variable**

```python
def func(x):
  print("Local x is",x)
  x=2
  print("Changed local x to:",x)
x=50
func(x)
print("x is still",x)
```

**Global variable**

```python
def func():
  global y
  print("y is",y)
  y=2
  print("y is changed to:",y)

y=50
func()
print("Value of y is",y)
```

```python
total=0
def sum(arg1,arg2):
  total=arg1+arg2
  print("Inside the function local total is",total)
  return total
sum(10,30)
print("Outside the function total is",total)
```

**Default argument**

```python
def func(a,b=10,c=5):
    print("a is",a,"b is",b,"c is",c)
func(3,b=8)# a is required, b is keyword, c is default
func(10,c=100)
func(25,24)
func(c=50,a=19)#parameters can be in any order
func(a=2) #a is required
```
