```python
import pandas as pd
production=pd.Series([30,60,40,55],
                     index=["Wheat","Rice","Maize","Ragi"])
print(production)
```

**loc--->location by label**

```python
print(production.loc["Rice"])
```

```python
print(production["Rice"])
```

**iloc--->location by indexing**

```python
print(production.iloc[2])
```

```python
print(production.sum())
print(production.mean())
print(production.max())
print(production.idxmax())
print(production.idxmax(),"is having maximum value of",production.max())
```

```python
increased_production=production*1.10
print("After increasing by 10%\n",increased_production)
```

```python
high_production=production[production>40]
print(high_production)
print()
print(production>40)
```

```python
import pandas as pd
data={
    "Farm":["Farm A","Farm B","Farm C","Farm D"],
    "Crops":["Wheat","Rice","Maize","Jowar"],
    "Area_hectares":[10,15,8,13],
    "Production_tons":[30,60,24,39]
    }

df=pd.DataFrame(data)
print(df)
```

**head --top rows
tail--bottom rows**

```python
print(df.head())
```

```python
print(df.head(3))
```

```python
print(df.tail())
```

```python
print(df.tail(2))
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
df["yield_ton_per_hectare"]=df["Production_tons"]/df["Area_hectares"]
print(df)
```

```python
print(df.tail(2))
```

```python
print(df.iloc[0:2])
```

```python
print(df.iloc[[1,3]])
```

```python
print(df.iloc[[1,3],[0,1,2]])
```

```python
print(df.loc[[1,3],["Farm","Crops","Area_hectares"]])
```

```python
print(df[1:4])
```

```python
wheat_farm=df[df["Crops"]=='Wheat']
print(wheat_farm)
```

```python
high_yield=df[df["yield_ton_per_hectare"]>3]
print(high_yield)
```

```python
sorted_df=df.sort_values(by="Production_tons",ascending=False)
print(sorted_df)
```

```python
avg_yield=df.groupby("Crops")["yield_ton_per_hectare"].mean()
print(avg_yield)
```
