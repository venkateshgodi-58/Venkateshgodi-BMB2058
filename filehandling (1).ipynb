```python
def expr(a,b):
  try:
    c=a/b

  except ZeroDivisionError:
    print("Divided by zero")
  except NameError:
    print("Variable not definied")
  except:
    print("Some issue ocured")
  else:
    print("No error occured")
    print("Value of c is",c)
  finally:
    print("End of the execution")
a=10
b=3
expr(a,b)
```
